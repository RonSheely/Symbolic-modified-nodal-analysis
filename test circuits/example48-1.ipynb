Solving the equations generated by node analysis for example48-1.net

```python
from sympy import *
import numpy as np
init_printing()
```

copy the symbols used in nodal analysis

```python
I_L1, s, I_V1, R3, R4, I_F1, R2, v3, v4, V2, ea1, I_V2, f1, v2, L1, V1, v1, I_Ea1, v5, I1 = symbols('I_L1 s I_V1 R3 R4 I_F1 R2 v3 v4 V2 ea1 I_V2 f1 v2 L1 V1 v1 I_Ea1 v5 I1')
```

copy the equation generated by nodal analysis

```python
equ = Matrix([[Eq(I_L1 + I_V1 + v1/R2 - v2/R2, 0)], [Eq(I_F1 + v2*(1/R4 + 1/R2) - v5/R4 - v1/R2, 0)], [Eq(I_Ea1 - I_F1 + v3/R3 - v4/R3, 0)], [Eq(-I_L1 - v3/R3 + v4/R3, -I1)], [Eq(-I_V2 - v2/R4 + v5/R4, 0)], [Eq(v1, V1)], [Eq(-v5, V2)], [Eq(-ea1*v1 + ea1*v4 + v3, 0)], [Eq(I_F1 - I_V2*f1 + v2 - v3, 0)], [Eq(I_L1/(L1*s) + v1 - v4, 0)]])
equ
```

```python
# enter the element values
equ1a = equ.subs({L1:2})

equ1a = equ1a.subs({s:1e10})
#equ1a = equ1a.subs({I_L1:0})
#equ1a = equ1a.subs({v1:2})
#equ1a = equ1a.subs({v5:0})

equ1a = equ1a.subs({R2:1})
equ1a = equ1a.subs({R3:1})
equ1a = equ1a.subs({R4:2})

equ1a = equ1a.subs({V1:2})
equ1a = equ1a.subs({V2:0})
equ1a = equ1a.subs({I1:9})
equ1a = equ1a.subs({f1:2})
equ1a = equ1a.subs({ea1:2})

equ1a  # display the equations
```

```python
str(solve(equ1a,[v1, v2, v3, v4, v5, I_V1, I_V2, I_Ea1, I_F1, I_L1]))
```
